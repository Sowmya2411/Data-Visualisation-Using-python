```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# date_range function from pandas by providing start date and periods
# periods = 500, no of periods (days) to generate
dates = pd.date_range('1/1/2000', periods=500)

# Let's generate a series using rand()
col_D = np.random.rand(500)

# Let's use numpy's randn for "standard normal" distribution.
#So if we plot hist from normal distribution, it would be bell shaped
data1 = np.random.randn(500,3) 

# randn(1000,4) 1000 by 4  2D array
df1 = pd.DataFrame(data = data1, 
                   index=dates, columns=['A', 'B', 'C'])
```

```python
#view of the dataframe df1
df1.head()
```

```python
#Adding col_D as a column to df1
df1['D']=col_D
```

```python
# rand(20,3) 20 by 3 2D array
data2 = np.random.rand(20,3)
col = ['X', 'Y', 'Z']
df2 = pd.DataFrame(data = data2, columns=col)
```

```python
df2.head()
```

```python
#Area charts depict a time-series relationship
#Area charts are perfect when communicating the overall trend, as opposed to the individual values.
#df2.plot.area(alpha=0.5)
df2.plot(kind = 'area', alpha=0.5)
```

```python
df1['A'].plot.hist(bins=30)
```

```python
#Line Plot
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
#To plot multiple column groups in a single axes, we need to repeat plot method specifying target ax.
#it is recommended to specify color and label keywords to distinguish each groups.
```

```python
ax = df1.plot.scatter(x='A',y='B', 
                 color='DarkBlue', label='Group 1')
df1.plot.scatter(x='C',y='D', 
                 color='DarkGreen', label='Group 2', ax = ax)
df2.plot.scatter(x='X',y='Y', color='red', label='Group 3', ax = ax)
```

```python
#cmap can be used to indicate colormap. 
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# We can use 's' to indicate size
df1.plot.scatter(x='A',y='B',c='C',s=50,cmap='coolwarm')
```

```python
#Boxplot plotting
df2.plot.box() 
```

```python
#Boxplot can be colorized by passing color keyword. 
color = dict(boxes='DarkGreen', whiskers='DarkOrange',
             medians='DarkBlue', caps='Gray')
```

```python
df2.plot.box(color=color)
```

```python
#Hexbin plots can be a useful alternative to scatter plots 
#if data is too dense to plot each point individually.
df1.plot.hexbin(x='A',y='B')#,gridsize=25,cmap='Oranges')
```

```python
#A larger gridsize means more,smaller bins.
df1.plot.hexbin(x='A',y='B',gridsize=10,cmap='Oranges')
```

```python
df1.plot.hexbin(x='A',y='B',gridsize=25,cmap='Oranges')
```

```python
#Kernel density plot
df2['X'].plot.kde()
```

```python
#multiple kde plots in the same figure
df2.plot.density()
```

```python
#pie Plots
df2.plot.pie(y='X')#,figsize=(12, 4),legend=False)
```
